```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import random
```

# Part 1

```python
# TODO define gaussian function here
def gaussian(x, sigma, amp):
    return amp*np.exp(-(x**2)/(2*sigma**2))
```

```python
# TODO plot gaussians here
xarr = np.arange(-5,5,0.1)

f, ax = plt.subplots(2, 2, sharex = True, sharey = True)
ax[0,0].plot(xarr, gaussian(xarr, 1, 1))
ax[0,1].plot(xarr, gaussian(xarr, 0.5, 2))
ax[1,0].plot(xarr, gaussian(xarr, 1, 2))
ax[1,1].plot(xarr, gaussian(xarr, 2, 1))
plt.show()
```

```python
# TODO fit gaussians, with known parameters, here
xarr = np.arange(-5,5,0.1)
yarr = gaussian(xarr,1,1)

params = scipy.optimize.curve_fit(gaussian, xarr, yarr)

# TODO plot result
plt.plot(xarr, gaussian(xarr,1,1), label = 'Original')
plt.plot(xarr, gaussian(xarr,params[0][0],params[0][1]), label = 'Fit', linestyle = '--')
plt.legend()
plt.show()
```

```python
# TODO: load data here
f = open("HDalpha_data", "r") 
data = f.read()

alt_data = data.split()

xarr = []
yarr = []
for e,i in enumerate(alt_data):
    i = float(i)
    if e%2 == 0:
        xarr.append(i)
    elif e%2 == 1:
        yarr.append(i)
```

```python
# TODO: plot it
plt.plot(xarr, yarr)
plt.title('Absorption')
plt.show()
```

```python
# TODO: plot data as emission signal
for i,e in enumerate(xarr):
    yarr[i] = -1*yarr[i]
   
plt.plot(xarr, yarr)
plt.title('Emission')
plt.show()
```

```python
# TODO: fit a function to this data and plot it
def double_gaussian(x, x1, x2, sigma1, sigma2, amp1, amp2):
    return (amp1*np.exp(-(np.power(x-x1,2))/(2*sigma1**2))) + (amp2*np.exp(-(np.power(x-x2,2))/(2*sigma2**2)))

# TODO fit gaussians, with known parameters, here
params = scipy.optimize.curve_fit(double_gaussian, xarr, yarr, p0 = [13128, 13132, 2, 2, 0.000025, 0.000023])

# TODO plot result
plt.title('Emission Spectrum of Hydrogen')
plt.plot(xarr, yarr, label = 'Original')
plt.plot(xarr, double_gaussian(xarr,params[0][0],params[0][1], params[0][2], params[0][3], params[0][4],params[0][5]), label = 'Fit', linestyle = '--')
plt.legend()
plt.show()
```

```python
# TODO: part 2
xarr = np.random.rand(100)
yarr = np.random.rand(100)

f, ax = plt.subplots(2, 2, sharex = True, sharey = True)
f.set_size_inches(8,8)

area1 = np.pi* (15*yarr[0:25])**2
colors1 = xarr[0:25]

area2 = np.pi* (15*xarr[25:50])**2
area3 = np.pi* (15*yarr[50:75])**2
area4 = np.pi* (15*xarr[75:100])**2

ax[0,0].tick_params(bottom="off", left="off")
ax[0,1].tick_params(bottom="off", left="off")
ax[1,0].tick_params(bottom="off", left="off")
ax[1,1].tick_params(bottom="off", left="off")

ax[0,0].scatter(xarr[0:25], yarr[0:25], c = colors1, cmap=plt.cm.gist_rainbow, alpha = 0.3, s = area1)
ax[0,1].scatter(xarr[25:50], yarr[25:50], color = 'r', alpha = 0.5, s = area2)
ax[1,0].scatter(xarr[50:75], yarr[50:75], color = 'g', alpha = 0.3, s = area3)
ax[1,1].scatter(xarr[75:100], yarr[75:100], color = 'b', alpha = 0.1, s = area4)

f.suptitle("Playing with Scatter Plots")
ax[0,0].set_title("Rainbow")
ax[1,0].set_title("Red")
ax[0,1].set_title("Green")
ax[1,1].set_title("Blue")

plt.show()

f.savefig('scatter.pdf')
```

# Part 3

```python
def flux_for_dipole(x_pos, y_pos, Q_vals, X, Y):
    """
    Returns the flux based on the information of the charges
    and the grid of the plot.
    Code modified from Physics 113, written by Blas Cabrera
    """
    flux = np.zeros((len(X), len(Y)))
    x_m = np.outer(np.ones((X.size)), X); 
    y_m = np.outer(Y, np.ones((Y.size))); 
    for n in range(len(Q_vals)):
        flux += 2*np.pi*Q_vals[n]*(1 - (x_pos[n]-x_m)/np.sqrt((y_pos[n]-y_m)**2+(x_pos[n]-x_m)**2))
    return flux
```

```python
# TODO create arrays of information
x_pos = np.random.rand(4)
y_pos = np.random.rand(4)

q_vals = np.random.randint(1,10,4)
sign = np.random.randint(0,2,4)

for i,e in enumerate(sign):
    if e == 0:
        q_vals[i] = -1*q_vals[i]
        
X = np.arange(-31.5, 32.5, 1)
Y = np.arange(-31.5, 32.5, 1)

# TODO get flux
flux = flux_for_dipole(x_pos, y_pos, q_vals, X, Y)

# TODO plot contours of the flux
plt.contour(X, Y, flux, 100)
plt.plot(x_pos, y_pos, 'o')
plt.show()
```

```python
x_pos = [1, 2]
y_pos = [1, 2]
q_vals = [-2, 2]
X = np.arange(-31.5, 32.5, 1)
Y = np.arange(-31.5, 32.5, 1)

flux = flux_for_dipole(x_pos, y_pos, q_vals, X, Y)

plt.contour(X, Y, flux, 200)
plt.plot(x_pos, y_pos, 'o')
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.show()
```
